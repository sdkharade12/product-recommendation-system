```python
# Extracting the text from description column

import ast

def clean_description(text):
    try:
        lst = ast.literal_eval(text)  # convert string → list
        if isinstance(lst, list):
            return ' '.join(lst)  # join list elements
        return text
    except:
        return ''

df1['description'] = df1['description'].apply(clean_description)
```

```python
# Extracting the text from category column
import ast

def clean_category(text):
    try:
        lst = ast.literal_eval(text)
        if isinstance(lst, list):
            return ' '.join(lst)
        return text
    except:
        return ''

df1['category'] = df1['category'].apply(clean_category)
```

```python
df1.head()
```

```python
df1['combined_text'] = (
    df1['title'] * 3 + " " +
    df1['description'] * 2 + " " +
    df1['category'] * 2 + " " +
    df1['brand']
)
```

```python
import re
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z0-9\s]', '', text)  # remove special chars
    return text
df1['combined_text'] = df1['combined_text'].apply(clean_text)
```

```python
df1.head()
```

```python
from sentence_transformers import SentenceTransformer

# Load the pre-trained model
print("Loading the embedding model...")
model = SentenceTransformer('all-MiniLM-L6-v2')
print("Model loaded successfully!")
```

```python
product_texts = df1['combined_text'].tolist()
```

```python
import os
import numpy as np

if os.path.exists("../models/product_embeddings.npy"):
    print("✅ Loading saved embeddings...")
    product_embeddings = np.load("../models/product_embeddings.npy")
else:
    print("⏳ Generating embeddings (first time only)...")
    product_embeddings = model.encode(
        product_texts,
        batch_size=64,
        show_progress_bar=True
    )
    np.save("../models/product_embeddings.npy", product_embeddings)
    print("✅ Embeddings saved!")
```

```python
#Building the Model
from sklearn.neighbors import NearestNeighbors

nn_model = NearestNeighbors(
    n_neighbors=6,      # 5 recommendations + 1 itself
    metric='cosine'
)

nn_model.fit(product_embeddings)
```

```python
def recommend_products(query, top_k=5):
    # Convert query → embedding
    query_embedding = model.encode([query])
    
    # Find nearest neighbors
    distances, indices = nn_model.kneighbors(query_embedding, n_neighbors=top_k)
    
    results = df1.iloc[indices[0]]
    
    return results[['title', 'brand', 'category']]
```

```python
recommend_products("bluetooth speaker")
```

```python
recommend_products("portable bluetooth speaker")
```

```python
recommend_products("sony bluetooth speaker")
```
