```python
df1 = df1.groupby('title').agg({
    'description': 'first',
    'category': 'first',
    'brand': 'first',
    'reviewText': ' '.join
}).reset_index()
```

```python
# Extracting the text from description column

import ast

def clean_description(text):
    try:
        lst = ast.literal_eval(text)  # convert string → list
        if isinstance(lst, list):
            return ' '.join(lst)  # join list elements
        return text
    except:
        return ''

df1['description'] = df1['description'].apply(clean_description)
```

```python
# Extracting the text from category column
import ast

def clean_category(text):
    try:
        lst = ast.literal_eval(text)
        if isinstance(lst, list):
            return ' '.join(lst)
        return text
    except:
        return ''

df1['category'] = df1['category'].apply(clean_category)
```

```python
df1.head()
```

```python
df1['combined_text'] = (
    df1['title'] + " " +
    df1['description'] + " " +
    df1['category'] + " " +
    df1['brand'] + " " +
    df1['reviewText']
)
```

```python
import re
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z0-9\s]', '', text)  # remove special chars
    return text
df1['combined_text'] = df1['combined_text'].apply(clean_text)
```
